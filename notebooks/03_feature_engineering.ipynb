# Feature Engineering
**Phase 2.5:** Create powerful derived metrics for SDG analysis.

**Objectives:**
1. Create Per Capita metrics (normalize by population)
2. Create Financial Ratios (Debt-to-GDP, Tax-to-GDP, etc.)
3. Calculate Year-over-Year Growth Rates
4. Create Composite Indicators (Fiscal Capacity, Social Spending Ratio, etc.)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Display settings
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)
sns.set_theme(style="whitegrid")
%matplotlib inline

print("âœ… Libraries loaded!")
```

## Load Cleaned Data

```python
# Load the cleaned data from Phase 2
df = pd.read_csv('../data/processed/fiscal_data_clean.csv')

print(f"Loaded data: {df.shape}")
print(f"Countries: {df['Country'].nunique()}")
print(f"Year range: {df['Year'].min():.0f} - {df['Year'].max():.0f}")
print("\nColumns:")
print(df.columns.tolist())
df.head()
```

## Feature Set 1: Per Capita Metrics (Normalize by Population)

```python
# Create Per Capita metrics (critical for fair country comparisons)
print("Creating Per Capita Metrics...")
print("="*60)

# Health Expenditure per Capita
if 'Health Expenditure' in df.columns and 'Population' in df.columns:
    df['Health_Expenditure_Per_Capita'] = df['Health Expenditure'] / df['Population']
    print("âœ… Health Expenditure per Capita")

# Education Expenditure per Capita
if 'Education Expenditure' in df.columns and 'Population' in df.columns:
    df['Education_Expenditure_Per_Capita'] = df['Education Expenditure'] / df['Population']
    print("âœ… Education Expenditure per Capita")

# Capital Expenditure per Capita (Infrastructure)
if 'Capital Expenditure' in df.columns and 'Population' in df.columns:
    df['Capital_Expenditure_Per_Capita'] = df['Capital Expenditure'] / df['Population']
    print("âœ… Capital Expenditure per Capita")

# Revenue per Capita
if 'Revenue' in df.columns and 'Population' in df.columns:
    df['Revenue_Per_Capita'] = df['Revenue'] / df['Population']
    print("âœ… Revenue per Capita")

# Tax Revenue per Capita
if 'Tax Revenue' in df.columns and 'Population' in df.columns:
    df['Tax_Revenue_Per_Capita'] = df['Tax Revenue'] / df['Population']
    print("âœ… Tax Revenue per Capita")

print(f"\n{'='*60}")
print(f"Per Capita features created: 5")
df[['Country', 'Year', 'Population', 'Health_Expenditure_Per_Capita', 'Revenue_Per_Capita']].head()
```

## Feature Set 2: Economic Ratios (The "Efficiency" Metrics)

```python
# Create Economic Ratios (Critical for governance analysis - SDG 16)
print("Creating Economic Ratios...")
print("="*60)

# 1. Debt-to-GDP Ratio (SDG 9: Infrastructure sustainability)
if 'Government Debt' in df.columns and 'Nominal GDP' in df.columns:
    df['Debt_to_GDP_Ratio'] = (df['Government Debt'] / df['Nominal GDP']) * 100
    print("âœ… Debt-to-GDP Ratio (%)")

# 2. Tax-to-GDP Ratio (SDG 16: Governance efficiency)
if 'Tax Revenue' in df.columns and 'Nominal GDP' in df.columns:
    df['Tax_to_GDP_Ratio'] = (df['Tax Revenue'] / df['Nominal GDP']) * 100
    print("âœ… Tax-to-GDP Ratio (%) - KEY GOVERNANCE METRIC")

# 3. Health Spending as % of GDP (SDG 3)
if 'Health Expenditure' in df.columns and 'Nominal GDP' in df.columns:
    df['Health_Spending_to_GDP'] = (df['Health Expenditure'] / df['Nominal GDP']) * 100
    print("âœ… Health Spending to GDP (%)")

# 4. Education Spending as % of GDP (SDG 4)
if 'Education Expenditure' in df.columns and 'Nominal GDP' in df.columns:
    df['Education_Spending_to_GDP'] = (df['Education Expenditure'] / df['Nominal GDP']) * 100
    print("âœ… Education Spending to GDP (%)")

# 5. Budget Balance as % of GDP (Fiscal health)
if 'Budget Deficit/Surplus' in df.columns and 'Nominal GDP' in df.columns:
    df['Budget_Balance_to_GDP'] = (df['Budget Deficit/Surplus'] / df['Nominal GDP']) * 100
    print("âœ… Budget Balance to GDP (%)")

# 6. Trade Balance (Exports - Imports)
if 'Exports' in df.columns and 'Imports' in df.columns:
    df['Trade_Balance'] = df['Exports'] - df['Imports']
    print("âœ… Trade Balance")

# 7. Trade Openness (Exports + Imports) / GDP
if 'Exports' in df.columns and 'Imports' in df.columns and 'Nominal GDP' in df.columns:
    df['Trade_Openness'] = ((df['Exports'] + df['Imports']) / df['Nominal GDP']) * 100
    print("âœ… Trade Openness (%)")

print(f"\n{'='*60}")
print(f"Economic Ratio features created: 7")
df[['Country', 'Year', 'Debt_to_GDP_Ratio', 'Tax_to_GDP_Ratio']].head()
```

## Feature Set 3: Growth Rates (Year-over-Year Changes)

```python
# Create Year-over-Year Growth Rates (Trend analysis)
print("Creating Growth Rate Metrics...")
print("="*60)

# Sort by Country and Year to ensure proper ordering
df = df.sort_values(['Country', 'Year']).reset_index(drop=True)

# Function to calculate YoY growth
def calculate_yoy_growth(group, column):
    """Calculate Year-over-Year percentage change"""
    return group[column].pct_change() * 100

# GDP Growth (already exists, but let's verify)
print("âœ… GDP Growth Rate (already in dataset)")

# Revenue Growth Rate
if 'Revenue' in df.columns:
    df['Revenue_Growth_Rate'] = df.groupby('Country')['Revenue'].transform(lambda x: x.pct_change() * 100)
    print("âœ… Revenue Growth Rate (YoY %)")

# Tax Revenue Growth Rate
if 'Tax Revenue' in df.columns:
    df['Tax_Revenue_Growth_Rate'] = df.groupby('Country')['Tax Revenue'].transform(lambda x: x.pct_change() * 100)
    print("âœ… Tax Revenue Growth Rate (YoY %)")

# Health Expenditure Growth Rate
if 'Health Expenditure' in df.columns:
    df['Health_Expenditure_Growth_Rate'] = df.groupby('Country')['Health Expenditure'].transform(lambda x: x.pct_change() * 100)
    print("âœ… Health Expenditure Growth Rate (YoY %)")

# Population Growth Rate
if 'Population' in df.columns:
    df['Population_Growth_Rate'] = df.groupby('Country')['Population'].transform(lambda x: x.pct_change() * 100)
    print("âœ… Population Growth Rate (YoY %)")

# Debt Growth Rate
if 'Government Debt' in df.columns:
    df['Debt_Growth_Rate'] = df.groupby('Country')['Government Debt'].transform(lambda x: x.pct_change() * 100)
    print("âœ… Debt Growth Rate (YoY %)")

print(f"\n{'='*60}")
print(f"Growth Rate features created: 6")
df[['Country', 'Year', 'Revenue_Growth_Rate', 'Tax_Revenue_Growth_Rate']].head(10)
```

## Feature Set 4: Composite Indicators (Advanced)

```python
# Create Composite/Advanced Indicators
print("Creating Composite Indicators...")
print("="*60)

# 1. Fiscal Capacity Index (Tax Revenue / Total Revenue)
if 'Tax Revenue' in df.columns and 'Revenue' in df.columns:
    df['Fiscal_Capacity_Index'] = (df['Tax Revenue'] / df['Revenue']) * 100
    print("âœ… Fiscal Capacity Index (% of revenue from taxes)")

# 2. Social Spending Ratio (Health + Education) / Total Expenditure
if 'Health Expenditure' in df.columns and 'Education Expenditure' in df.columns and 'Expenditure' in df.columns:
    df['Social_Spending_Ratio'] = ((df['Health Expenditure'] + df['Education Expenditure']) / df['Expenditure']) * 100
    print("âœ… Social Spending Ratio (% of budget on health + education)")

# 3. Infrastructure Investment Ratio (Capital Expenditure / GDP)
if 'Capital Expenditure' in df.columns and 'Nominal GDP' in df.columns:
    df['Infrastructure_Investment_Ratio'] = (df['Capital Expenditure'] / df['Nominal GDP']) * 100
    print("âœ… Infrastructure Investment Ratio (% of GDP)")

# 4. Economic Diversification (simple proxy: exports as % of GDP)
if 'Exports' in df.columns and 'Nominal GDP' in df.columns:
    df['Export_Intensity'] = (df['Exports'] / df['Nominal GDP']) * 100
    print("âœ… Export Intensity (% of GDP)")

# 5. Inflation Volatility (rolling 3-year std dev)
if 'Inflation Rate' in df.columns:
    df['Inflation_Volatility'] = df.groupby('Country')['Inflation Rate'].transform(lambda x: x.rolling(3, min_periods=1).std())
    print("âœ… Inflation Volatility (3-year rolling std)")

print(f"\n{'='*60}")
print(f"Composite features created: 5")
df[['Country', 'Year', 'Fiscal_Capacity_Index', 'Social_Spending_Ratio']].head()
```

## Feature Summary & SDG Mapping

```python
# Summary of all engineered features
print("="*70)
print("FEATURE ENGINEERING SUMMARY")
print("="*70)

# List all new features
original_cols = pd.read_csv('../data/processed/fiscal_data_clean.csv').columns
new_features = [col for col in df.columns if col not in original_cols]

print(f"\nðŸ“Š Total New Features Created: {len(new_features)}")
print("\nFeature Categories:")
print(f"  - Per Capita Metrics: 5")
print(f"  - Economic Ratios: 7")
print(f"  - Growth Rates: 6")
print(f"  - Composite Indicators: 5")
print(f"  - TOTAL: 23 engineered features")

print("\n" + "="*70)
print("SDG-CRITICAL FEATURES (For Our Narrative)")
print("="*70)

sdg_features = {
    'SDG 8 (Economic Growth)': [
        'GDP Growth Rate',
        'Tax_to_GDP_Ratio',
        'Revenue_Growth_Rate',
        'Tax_Revenue_Growth_Rate',
        'Fiscal_Capacity_Index',
        'Trade_Balance',
        'Unemployment Rate'
    ],
    'SDG 16 (Governance)': [
        'Tax_to_GDP_Ratio',
        'Fiscal_Capacity_Index',
        'Inflation_Volatility',
        'Inflation Rate'
    ],
    'SDG 9 (Infrastructure)': [
        'Debt_to_GDP_Ratio',
        'Infrastructure_Investment_Ratio',
        'Capital_Expenditure_Per_Capita',
        'Debt_Growth_Rate'
    ],
    'SDG 3 (Health)': [
        'Health_Expenditure_Per_Capita',
        'Health_Spending_to_GDP',
        'Health_Expenditure_Growth_Rate'
    ],
    'SDG 1 (Poverty)': [
        'GDP per Capita',
        'Real GDP',
        'Revenue_Per_Capita'
    ]
}

for sdg, features in sdg_features.items():
    print(f"\n{sdg}:")
    for feature in features:
        if feature in df.columns:
            print(f"  âœ… {feature}")
        else:
            print(f"  âš ï¸ {feature} (not available)")
```

## Data Quality Check

```python
# Check for any issues with engineered features
print("Data Quality Check on Engineered Features:")
print("="*60)

# Check for infinite values (division by zero)
inf_cols = []
for col in new_features:
    if df[col].dtype in ['float64', 'int64']:
        inf_count = np.isinf(df[col]).sum()
        if inf_count > 0:
            inf_cols.append((col, inf_count))
            # Replace inf with NaN
            df[col] = df[col].replace([np.inf, -np.inf], np.nan)

if inf_cols:
    print("\nâš ï¸ Found and fixed infinite values:")
    for col, count in inf_cols:
        print(f"  - {col}: {count} infinite values â†’ converted to NaN")
else:
    print("âœ… No infinite values found")

# Check missing values in new features
print("\n" + "="*60)
print("Missing Values in New Features:")
missing_new = df[new_features].isnull().sum()
missing_new = missing_new[missing_new > 0].sort_values(ascending=False)
if len(missing_new) > 0:
    print(missing_new)
else:
    print("âœ… No missing values in engineered features!")

# Statistical summary
print("\n" + "="*60)
print("Statistical Summary (Sample Features):")
print(df[['Debt_to_GDP_Ratio', 'Tax_to_GDP_Ratio', 'Health_Expenditure_Per_Capita']].describe())
```
